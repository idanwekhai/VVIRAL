```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
AAV2_base = [78,48.9,52.63,42.07,52.63,63.16,43.82,43.75,30.1,31.5,39.04,53.34,35.87,32]
AAV9_base = [4.8, 3.77, 29.36, 1.06, 0.39, 12.35, 9.27, 2.28, 24.77, 6.48]
AAV6_base = [49.98, 46.35, 50.81, 44.35, 42.54, 37.75, 33.04, 32.98, 67.34, 15.03]
AAV5_base = [19.17, 45.83, 43.40, 16.32 ,29.91, 25, 10.61, 29.46, 6.5, 8.09, 14.94, 0.86]
```

```python
AAV2 = pd.read_csv("../data/SEC/AAV2/yields.csv").values[:,:3].T.tolist()
AAV9 = pd.read_csv("../data/SEC/AAV9/yields.csv").values[:,:3].T.tolist()
AAV6 = []
AAV5 = pd.read_csv("../data/SEC/AAV5/yields.csv").values[:,:].T.tolist()
```

```python
AAV2.insert(0, AAV2_base)
AAV9.insert(0, AAV9_base)
AAV6.insert(0, AAV6_base)
AAV5.insert(0, AAV5_base)
```

```python
categories = ['Baseline', 'Batch 1', 'Batch 2', 'Batch 3']
num_subplots = 4 


# Create a figure and axes
fig, axes = plt.subplots(nrows=1, ncols=num_subplots, figsize=(9, 3.5), sharey=True)

AAVs = ['AAV2', 'AAV5', 'AAV6', 'AAV9']
data = [AAV2, AAV5, AAV6, AAV9]
# Plot each category in a subplot
for i in range(4):
    ax = axes[i]
    sns.boxplot(data=data[i], ax=ax, width=0.4, saturation=1, palette='pastel')
    
    ax.set_xticklabels(categories, rotation=20)
    ax.set_title(AAVs[i], fontdict={'size':12, 'weight': 'bold'})

    if i == 0:
        ax.set_ylabel('Yield (%)', fontdict={'size':12, 'weight': 'bold'})


plt.tight_layout()
# plt.show()
plt.savefig('SEC_boxplot_yield_max_saturation.png', dpi=400, bbox_inches='tight', transparent=True)
```
